```python
# open files and convert to ordered dict
# I am defining a variable with my file name
filepath = 'data/'
datalogger_file = 'monkeys_80-30_dr20_try3_2020-03-03_170333_temp2'
# avg_export_name = filepath +'summary_' + '2S-18g' + '_20-PPMLPM_NO2_' + '20190923' +'.csv'

# nameoftest = filename[2:5]+' PPM NO2'
# test_date = filename[6:16]
# avg_export_name = filepath +'summary_' + filename + '_PPM_NO2_' + test_date +'.csv'
#avg_export_name = filepath +'summary_' + filename[2:5] + '_PPM_NO2_' + test_date +'.csv'
# print (nameoftest, avg_export_name, test_date)

# Loads the specified datalogger file into a python dictionary (array indexed with strings)
# datalogger file from control PC with data for MFC's, etc.
data = csv_to_data.loadCSV(filepath + datalogger_file + '.csv') #datalogger file
tm = data['elapsed_time']

#in case a GSAT was used the following statement will look for GSAT data when CAPS data isn't found.
# if not 'CAPS.NO2' in data:
#     try:
#         data2 = csv_to_data.loadCSV(filepath + GSAT_file + '.csv') #GSAT data
#         data2['timestamp'] = data2['Timestamp']
#         print('data2 keys are: ',data2.keys())
#         data = csv_to_data.mergeAndTimeCorrelateData(data, data2)
#         data['CAPS.NO2'] = np.empty(len(data['elapsed_time']))
#         data['CAPS.NO2'] = data['NO2'] #convert gsat NO2 to caps key for processing.
#     except:
#         pass
# if not 'MFC1.massflow' in data:
#     try:
#         data['MFC1.mass_flow'] = data['MFC2.mass_flow']
#     except:
#         pass
# if not 'IGTH.humid' in data:
#     try:
#         data['IGTH.humid'] = numpy.zeros(len(tm))
#         data['IGTH.temp'] = numpy.zeros(len(tm))
#     except:
#         pass
data['prescrubber_valve_total'] = data['valve_ctrl.valve0'] + data['valve_ctrl.valve1'] + data['valve_ctrl.valve2'] + data['valve_ctrl.valve3']
data['postscrubber_valve_total'] = data['valve_ctrl.valve4'] + data['valve_ctrl.valve5'] + data['valve_ctrl.valve6'] + data['valve_ctrl.valve7']
data['valve_total'] = data['prescrubber_valve_total'] + data['postscrubber_valve_total']
# print(max(data['elapsed_time'])/3600, max(tm))

data['prescrubber_sample'] = numpy.empty(len(tm),dtype=int)
data['postscrubber_sample'] = numpy.empty(len(tm),dtype=int)
data['valve_sampled'] = numpy.empty(len(tm),dtype=int)


data_proc = {}
data_proc['start_sample_v1'] = numpy.empty(len(tm)+1,dtype=int)
data_proc['end_sample_v1'] = numpy.empty(len(tm)+1,dtype=int)
print(data.keys())
print(data_proc.keys())
```

```python
# Constants
total_num_valves = 8
total_num_scrubbers = 4
time_step = (tm[20]-tm[10])/10 #figure out sample rate
```

```python
# prescrubber = 0
# postscrubber = 0
for j in range(len(tm)):
    if int(data['prescrubber_valve_total'][j]) == total_num_valves/2 - 1: #check to see if sampling pre scrubber
        data['prescrubber_sample'][j] = 1
    else:
        data['prescrubber_sample'][j] = 0
    
    if int(data['postscrubber_valve_total'][j]) == total_num_valves/2 - 1: #check to see if sampling post scrubber
        data['postscrubber_sample'][j] = 1
    else:
        data['postscrubber_sample'][j] = 0
    
    if data['valve_total'][j] == total_num_valves: #check to see if the system is flushing
        data['valve_sampled'][j] = total_num_valves + 2 #set valve state beyond total number of valves
    #set  valve_sampled equal to the number of the valve sampled
    elif data['prescrubber_sample'][j] == 1: #if sampling pre scrubber
        if int(data['valve_ctrl.valve0'][j]) ==  0:
            data['valve_sampled'][j] = 0
        elif int(data['valve_ctrl.valve1'][j]) ==  0:
            data['valve_sampled'][j] = 1 
        elif int(data['valve_ctrl.valve2'][j]) ==  0:
            data['valve_sampled'][j] = 2 
        elif int(data['valve_ctrl.valve3'][j]) ==  0:
            data['valve_sampled'][j] = 3 
    elif data['postscrubber_sample'][j] == 1: #if sampling pre scrubber
        if int(data['valve_ctrl.valve4'][j]) ==  0:
            data['valve_sampled'][j] = 4
        elif int(data['valve_ctrl.valve5'][j]) ==  0:
            data['valve_sampled'][j] = 5 
        elif int(data['valve_ctrl.valve6'][j]) ==  0:
            data['valve_sampled'][j] = 6 
        elif int(data['valve_ctrl.valve7'][j]) ==  0:
            data['valve_sampled'][j] = 7  
    
    
#     if data['valve_ctrl.valve0'][j] == 0:
#         sample_time[i] += 1
    
        
```
